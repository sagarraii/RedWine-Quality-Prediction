```python
from dataclasses import dataclass
from pathlib import Path

@dataclass
class DataValidationConfig:
    root_dir: Path
    unzipped_data_dir: Path
    STATUS_FILE: str
    all_schema: dict
```

```python
from src.datascience.constants import *
from src.datascience.utils.common import read_yaml, create_directories
```

```python
class ConfigurationManager:
    def __init__(self,
                 config_filepath=CONFIG_FILE_PATH,
                 params_filepath = PARAMS_FILE_PATH,
                 schema_filepath = SCHEMA_FILE_PATH):
        self.config=read_yaml(config_filepath)
        self.params=read_yaml(params_filepath)
        self.schema=read_yaml(schema_filepath)

        create_directories([self.config.artifacts_root])

    def get_data_validation_config(self) -> DataValidationConfig:
        config = self.config.data_validation
        schema = self.schema.COLUMNS

        create_directories([config.root_dir])

        data_validation_config = DataValidationConfig(
            root_dir = config.root_dir,
            STATUS_FILE = config.STATUS_FILE,
            unzipped_data_dir = config.unzipped_data_dir,
            all_schema = schema 
        )

        return data_validation_config
```

```python
class DataValidation:
    def __init__(
            self,
            config = DataValidationConfig
    ):
        self.config = config


    def validate_all_columns(self) -> bool:
        try:
            validation_status = None

            df = pd.read_csv(self.config.unzipped_data_dir)
            cols_list = list(df.columns)

            all_schema = self.config.all_schema

            for cols, dtype in all_schema.items():
                if cols not in cols_list:
                    validation_status = False

                    with open(self.config.STATUS_FILE, 'w') as f:
                        f.write(f"Validation Status: {validation_status}")
                        
                elif cols_list.dtype != dtype:
                    validation_status = True

                    with open(self.config.STATUS_FILE, 'w') as f:
                        f.write(f"Validation Status: {validation_status}")

            return validation_status
        
        except Exception as e:
            raise e
```

```python
class DataValidation:
    def __init__(self, config):
        self.config = config

    def validate_all_columns(self) -> bool:
        try:
            validation_status = True

            df = pd.read_csv(self.config.unzipped_data_dir)

            # schema from yaml/config
            # example:
            # all_schema:
            #   fixed acidity: float
            #   quality: int

            all_schema = self.config.all_schema

            # Check columns + datatype
            for col, dtype in all_schema.items():

                # check column exists
                if col not in df.columns:
                    print(f"Missing column: {col}")
                    validation_status = False

                # check datatype
                elif df[col].dtype != dtype:
                    print(f"Datatype mismatch in {col}")
                    validation_status = False

            # check extra columns
            for col in df.columns:
                if col not in all_schema:
                    print(f"Unexpected column: {col}")
                    validation_status = False

            with open(self.config.STATUS_FILE, 'w') as f:
                f.write(f"Validation Status: {validation_status}")

            return validation_status

        except Exception as e:
            raise e
```
